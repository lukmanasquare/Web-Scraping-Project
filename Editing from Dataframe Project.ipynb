```python
import pandas as pd
```

```python
df= pd.read_csv("home_and_design.csv", encoding= 'latin-1')
df
```

```python
df.isna().sum()
```

```python
#to get the exact row where "name of business" column is null
nullbiz= df.loc[df['name of business'].isna()]
nullbiz
```

```python
df['name of business'] = df['name of business'].str.replace(' Black-Owned Brooklyn', '').str.strip()
df['name of business']
```

```python
#remove any closed business
#covert name of business column to str
df['name of business'] = df['name of business'].astype(str)

# create a boolean maskwhere the "name of business" column contains "(Closed)"
mask= df['name of business'].str.contains('(Closed)')

#use the .loc[] to select the rows that doesn't contain the string  "(Closed)"
#the ~ gives you the inverse 'mask' i.e it locates rows without the str "(Closed)" in the business name
df = df.loc[~mask]
df
```

```python
df['business category'] = 'Home and Design'
df
```

```python
df.to_csv("home_and_design.csv")
```
